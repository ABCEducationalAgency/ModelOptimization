```python
import tensorflow as tf
import os

model = tf.keras.models.load_model("mobilenetv2Model.keras")

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # Full quantization
tflite_model = converter.convert()

with open("mobilenetv2ModelCompressed.tflite", "wb") as f:
    f.write(tflite_model)


size_mb = os.path.getsize("mobilenetv2ModelCompressed.tflite") / (1024 * 1024)
print(f"Compressed Model Size: {size_mb:.2f} MB")
```
